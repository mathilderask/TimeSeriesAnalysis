```r
files <- dir("functions",full.names=TRUE)
for(i in 1:length(files)) source(files[i])
X <- read.table("box_data_60min.csv", sep=",", header=TRUE)
X$thour <- X$thour - X$thour[1] + 1
teststart = 167
Xtrain <- X[X$thour < teststart, ]
Xtest <- X[X$thour >= teststart, ]
```

```r
par(bg="white")

plot(X$thour, X$Ph, type="l", main="Heat from electrical heaters vs Time", 
     xlab="Hours (h)", ylab="Power (W)", cex.main=1.5, cex.lab=1.5, cex.axis=1.5)

plot(X$thour, X$Gv, type="l", main="Vertical solar radiation vs Time", 
     xlab="Hours (h)", ylab="Radiation (W/m^2)", cex.main=1.5, cex.lab=1.5, cex.axis=1.5)

plot(X$thour, X$Tdelta, type="l", main="Difference internal and external temperature vs Time", 
     xlab="Hours (h)", ylab="Diff Temperature (C)", cex.main=1.5, cex.lab=1.5, cex.axis=1.5)
```
