```r
phi1 = -0.7
phi2 = -0.2

# list with 30 nan values
p <- rep(NA, 30)


p[1] <- -phi1/(1+phi2)
p[2] <- phi1**2/(1+phi2)-phi2

for (k in 3:30){
    p[k] <- -phi1*p[k-1]-phi2*p[k-2]
}
```

```r
par(bg = "white")
k = seq(1,30,1)
plot(c(0,k),c(1,p[k]), xlab = "lag k",ylab = "p(k)",
main = "Autocorrelation function the AR(2) process", 
lwd = 2,cex = 1.5,cex.axis = 1.5,cex.lab = 1.5, cex.main = 1.5)
#lines from the x axis to the points
segments(0,0,0,1,lwd = 2)
for (k in 1:30){
    segments(k,0,k,p[k],lwd = 2)
}


# font si
```
