```python
## ---------------------------------------------------------------------------------------
# Write an ARMA simulation
n <- 200
y <- numeric(n)
# AR1: 
# (1 - phi1 * B) * Y_t = eps_t
# 



## ---------------------------------------------------------------------------------------
# Function for plot
plotit <- function(x){
  layout(rbind(1,2:3))
  par(mar=c(3,3,1,1), mgp=c(2, 0.7,0),bg="white")
  plot(x, ylab="X")
  acf(x, lag.max=50, lwd=2)
  pacf(x, lag.max=50, lwd=2)
}
```

# $A(1,0,0)X(0,0,0)_{12}$

```python
# AR(1)
n <- 1000
plotit(arima.sim(model=list(ar=c(-0.6)), n=n) ) # Hvorfor er phi -0.6 og ikke 0.6?
```

```python
plotit(arima.sim(model=list(ar=c(rep(0,11),0.9)), n=n))
```
