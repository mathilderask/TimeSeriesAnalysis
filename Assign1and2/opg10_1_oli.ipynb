```r
# Predictions in a state space model
# defince a function to predict the next state

model <- function(t) {
  # Preallocate vectors
  Xi <- numeric(t)
  Xm <- numeric(t)
  
  # Initial values
  Xi[1] <- -3.0
  Xm[1] <- -1.2
  U <- 2
  
  for (i in 2:t) { #“Start a loop where i goes from 2 up to t, one by one.
    eps_i <- rnorm(1, mean = 0, sd = 0.04)
    eps_m <- rnorm(1, mean = 0, sd = 0.01)
    
    Xi[i] <- 0.7 * Xi[i - 1] + 0.2 * Xm[i - 1] + 0.8 * U + eps_i
    Xm[i] <- 0.04 * Xi[i - 1] + 0.95 * Xm[i - 1] + 0.02 * U + eps_m
  }
  
  # Return as data frame
  return(data.frame(time = 1:t, Xi = Xi, Xm = Xm))
}

model(4)
```
