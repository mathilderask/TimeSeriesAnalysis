```r
#Question 1 Predict the oxygen content and phosphate concentration for the
#next two weeks
x1 <- c(c(3.1, 3.2, 2.9, 0.2, -2.3, -3.4),rep(0,2))
x2 <- c(c(2.1, 1.6, 1.8, 4.9, 3.5, 1.8),rep(0,2))
e1 <- rep(0,8)
e2 <- rep(0,8)

# get epsilin
for (i in 4:6){
    e1[i] <- x1[i]-0.8*x1[i-1] + 0.6*x2[i-1]
    e2[i] <- x2[i]-0.4*x1[i-1] - 0.7*x2[i-1] + 0.4*e2[i-1]
}

# prediction for the next two weeks with kalman filter

for (i in 7:8) {
    x1[i] <- 0.8*x1[i-1] - 0.6*x2[i-1] 
    x2[i] <- 0.4*x1[i-1] + 0.7*x2[i-1] - 0.4*e2[i-1]
}
par(bg = "lightblue")
plot(x1, type = "o", col = "red", ylim = c(-5, 5), ylab = "Oxygen content", xlab = "Time")
points(x2, type = "o", col = "blue")

x1
```
